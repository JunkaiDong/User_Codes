```python
# Load scikit learn standard datasets
from sklearn import datasets

# Import plot module
import matplotlib.pyplot as plt

# Import Pandas dataframe module
import pandas as pd

# Import scikit learn K-Means clustering module
from sklearn.cluster import KMeans
```

```python
# Read dataset to pandas dataframe
iris = datasets.load_iris()
```

```python
# Features and target
x = iris.data[:, :2]
y = iris.target
```

```python
# Plot data
def plot_clusters(x_arr, y_arr, xtitle="", ytitle="", title=None):    
    plt.scatter(x_arr[:,0], x_arr[:,1], c=y_arr)
    plt.xlabel(xtitle, fontsize=18)
    plt.ylabel(ytitle, fontsize=18)
    if not title:
        title = ''.join([xtitle, " vs. ", ytitle])
    plt.title(title, fontsize=18)
    plt.show()
    
plot_clusters(x, y, xtitle="Spea1_Length", ytitle="Sepal_Width", title="Actual Clusters")
```

```python
# K-Means results
kmeans_obj = KMeans(n_clusters = 3, n_jobs = 4, random_state=21)
kmeans_obj.fit(x)
```

```python
# Get clusters' centers
print(kmeans_obj.cluster_centers_)
```

```python
new_labels = kmeans_obj.labels_

plot_clusters(x, new_labels, xtitle="Spea1_Length", ytitle="Sepal_Width", title="K-Means Predicted Clusters")
```
